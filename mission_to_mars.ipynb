# NASA Mars News

```python
# PREP THE SOUP #
url = "https://mars.nasa.gov/news/"
#url = 'https://mars.nasa.gov/news/?page=0&per_page=40&order=publish_date+desc%2Ccreated_at+desc&search=&category=19%2C165%2C184%2C204&blank_scope=Latest'

browser =  Browser('chrome', **executable_path, headless=False)
browser.visit(url)
html = browser.html
soup = bs(html, 'html.parser')

# GET HTML DATA #
comment='''
get title and first paragraph

<li class='slide'> - article panel
to use:
<div class='list_text'> - text area of panel
<div class='content_title'>.a.text - for title
<div class='article_teaser_body'> - for first paragraph
'''
results = soup.find_all('li', class_='slide')

for res in results:
    title = res.find('div', class_='content_title').text
    date = res.find('div', class_='list_date').text
    teaser = res.find('div', class_='article_teaser_body').text
    print(date)
    print(title.upper())
    print(teaser)
    print("------------------------------\n")

browser.quit()
```
